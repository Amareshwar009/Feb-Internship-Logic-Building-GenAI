## 1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as different words because computers are case-sensitive. Uppercase and lowercase letters are considered different.

For example:
"Love is important. I love coding."

Without converting text to lowercase, "Love" and "love" are counted separately. NLP systems use case normalization to convert all text into lowercase.


## 2. What happens if stopwords are not removed?

If stopwords are not removed:

- The dataset becomes larger
- Processing becomes slower
- Important words may get hidden
- Model accuracy may decrease

Stopwords like "is", "the", and "in" do not add much meaning.


## 3. Provide two real-world scenarios where removing stopwords can be harmful.

### Scenario 1: Sentiment Analysis

Sentence:
"This product is not good."

If "not" is removed:
"This product good."

This changes the meaning from negative to positive.


### Scenario 2: Medical Records

Sentence:
"Patient does not have fever."

Removing "not" changes meaning and may cause incorrect medical decisions.


## 4. Explain the difference between stemming and lemmatization.

Stemming removes word endings using simple rules.

Examples:
playing → play  
studies → studi  

It may produce incorrect words.

Lemmatization converts words into dictionary form.

Examples:
playing → play  
studies → study  
better → good  

Lemmatization is more accurate but slower than stemming.

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []
clean_sentences = []

for sentence in test_sentences:

    tokens, clean_sentence = preprocess_text(sentence)

    print("\nOriginal Text:")
    print(sentence)

    print("Cleaned Tokens:")
    print(tokens)

    print("Cleaned Sentence:")
    print(clean_sentence)

    all_tokens.extend(tokens)
    clean_sentences.append(clean_sentence)
```

```python
print("\nToken Analytics:\n")

for tokens in all_tokens:

    if tokens:

        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))

        avg_length = round(
            sum(len(word) for word in tokens) /
            total_tokens, 2
        )

        print({
            "total_tokens": total_tokens,
            "unique_tokens": unique_tokens,
            "avg_token_length": avg_length
        })
```

```python
from collections import Counter

word_freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

```python
def full_pipeline(text_list):

    tokens_all = []
    clean_sentences_all = []

    for text in text_list:

        tokens, sentence = preprocess_text(text)

        tokens_all.extend(tokens)
        clean_sentences_all.append(sentence)

    return {
        "tokens": tokens_all,
        "clean_sentences": clean_sentences_all
    }

# Run Pipeline
pipeline_output = full_pipeline(test_sentences)

print(pipeline_output)
```
